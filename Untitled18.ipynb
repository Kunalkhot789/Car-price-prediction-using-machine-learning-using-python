```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe_brand= OneHotEncoder(sparse_output = False)
brand= ohe_brand.fit_transform (df[['brand']])
brand
```

```python
brand_df = pd.DataFrame(brand, columns=ohe_brand.categories_ )
```

```python
df['type'].value_counts()
```

```python
ohe_type = OneHotEncoder(sparse_output = False)
type = ohe_type.fit_transform(df[['type']])
type
```

```python
type_df = pd.DataFrame(type, columns=ohe_type.categories_)
type_df.drop('Convertible', axis=1, inplace=True)
type_df
```

```python
df = pd.concat((df, type_df, model_df, brand_df), axis =1)
```

```python
df.drop(['brand', 'model', 'type'], axis=1, inplace=True)
```

```python
df
```

```python
x_df = df.drop(['msrp'], axis=1)
```

```python
y = df['msrp']
```

```python
x = x_df.values
```

```python
y = y.values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
y_train_prede = lr_model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse_t = mean_squared_error(y_train_prede, y_train)
rmse_t = mean_squared_error(y_train_prede , y_train) ** 0.5
mse_t = mean_absolute_error(y_train_prede, y_train)
r2_train = r2_score(y_train, y_train_prede)


print(mse_t)
print(rmse_t)
print(mse_t)
print(r2_train)
```

```python
y_test_prede = lr_model.predict(x_test)
```

```python
mse_test = mean_squared_error(y_test_prede, y_test)
rmse_test = mean_squared_error(y_test_prede , y_test) ** 0.5
mse_test = mean_absolute_error(y_test_prede, y_test)
r2_test = r2_score(y_test, y_test_prede)


print(mse_test)
print(rmse_test)
print(mse_test)
print(r2_test)
```
